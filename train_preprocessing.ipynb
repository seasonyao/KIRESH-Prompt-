### KIRESH_baseline_data

```python
Community_Present_input = []
Community_Present_label = []

for i in range(df.shape[0]):
    input_sent = df.iloc[i]['SOCIAL_HISTORY'] 
    Community_Present_input.append(input_sent)
    Community_Present_label.append(df.iloc[i]['sdoh_community_present'])
    
Community_Absent_input = []
Community_Absent_label = []

for i in range(df.shape[0]):
    input_sent = df.iloc[i]['SOCIAL_HISTORY']        
    Community_Absent_input.append(input_sent)
    Community_Absent_label.append(df.iloc[i]['sdoh_community_absent'])
    
Education_input = []
Education_label = []

for i in range(df.shape[0]):
    input_sent = df.iloc[i]['SOCIAL_HISTORY']        
    Education_input.append(input_sent)
    Education_label.append(df.iloc[i]['sdoh_education'])
    
Economics_input = []
Economics_label = []

for i in range(df.shape[0]):
    input_sent = df.iloc[i]['SOCIAL_HISTORY']        
    Economics_input.append(input_sent)
    Economics_label.append(df.iloc[i]['sdoh_economics'])
    
Environment_input = []
Environment_label = []

for i in range(df.shape[0]):
    input_sent = df.iloc[i]['SOCIAL_HISTORY']        
    Environment_input.append(input_sent)
    Environment_label.append(df.iloc[i]['sdoh_environment'])
    
Alcohol_input = []
Alcohol_label = []

for i in range(df.shape[0]):
    input_sent = df.iloc[i]['SOCIAL_HISTORY']        
    Alcohol_input.append(input_sent)
    Alcohol_label.append(df.iloc[i]['behavior_alcohol'])
    
Tobacco_input = []
Tobacco_label = []

for i in range(df.shape[0]):
    input_sent = df.iloc[i]['SOCIAL_HISTORY']        
    Tobacco_input.append(input_sent)
    Tobacco_label.append(df.iloc[i]['behavior_tobacco'])
    
Drug_input = []
Drug_label = []

for i in range(df.shape[0]):
    input_sent = df.iloc[i]['SOCIAL_HISTORY']        
    Drug_input.append(input_sent)
    Drug_label.append(df.iloc[i]['behavior_drug'])
    
df_Community_Present = pd.DataFrame({'input': Community_Present_input, 'label': Community_Present_label})
df_Community_Absent = pd.DataFrame({'input': Community_Absent_input, 'label': Community_Absent_label})
df_Education = pd.DataFrame({'input': Education_input, 'label': Education_label})
df_Economics = pd.DataFrame({'input': Economics_input, 'label': Economics_label})
df_Environment = pd.DataFrame({'input': Environment_input, 'label': Environment_label})
df_Alcohol = pd.DataFrame({'input': Alcohol_input, 'label': Alcohol_label})
df_Tobacco = pd.DataFrame({'input': Tobacco_input, 'label': Tobacco_label})
df_Drug = pd.DataFrame({'input': Drug_input, 'label': Drug_label})

df_Community_Present.to_csv("KIRESH_baseline_data/community_present_train.csv")
df_Community_Absent.to_csv("KIRESH_baseline_data/community_absent_train.csv")
df_Education.to_csv("KIRESH_baseline_data/education_train.csv")
df_Economics.to_csv("KIRESH_baseline_data/economics_train.csv")
df_Environment.to_csv("KIRESH_baseline_data/environment_train.csv")
df_Alcohol.to_csv("KIRESH_baseline_data/alcohol_train.csv")
df_Tobacco.to_csv("KIRESH_baseline_data/tobacco_train.csv")
df_Drug.to_csv("KIRESH_baseline_data/drug_train.csv")
```
